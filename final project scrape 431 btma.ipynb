```python
from urllib.request import urlopen
from bs4 import BeautifulSoup 

import datetime
import re

import pandas as pd
import numpy as np
```

```python
def find_page_number(url):
    
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    try:
        View_page = bs.find('div',class_='showing').text
        number_of_pages = View_page.split()[-2].replace(',','')
        number_of_pages = int(int(number_of_pages)/40)
    except:
        number_of_pages = 1
        pass
    
    return number_of_pages  
```

```python
def get_item_info(url):
    every_element_list = []
    
    
    brand = ''
    item_carfax_link = ''
    item_dealer_add = ''
    Model = ''
    date = ''
    item_price = ''
    colour = ''
    configuration = ''
    item_condition = ''
    item_bodytype = ''
    item_wheelConfig = ''
    item_transmission = ''
    item_fueltype = ''
    item_mileage = ''

   
        
    html = urlopen(url)
    bs_item = BeautifulSoup(html, 'html.parser')


    try: 
        brand = bs_item.find(itemprop='brand').text
    except:
    
        brand =''
    
 
    try:
        Model = bs_item.find(itemprop='model').text
    except:
        Model =''

    try:
        item_fueltype = bs_item.find(itemprop='fuelType').text
    except:
        item_fueltype = ''


    try:
        date = bs_item.find(itemprop='vehicleModelDate').text
        date = int(date)
    except:
        date = 0


    try:
        item_condition = bs_item.find(itemprop='itemCondition').text
    except:
        item_condition = ''

    try:
        item_price = bs_item.find('span',itemprop='price').text
        item_price = int(item_price.replace('$','').replace(',','').replace('.',' ').split()[0])
    except:
        item_price = 0

    try:
        colour = bs_item.find(itemprop='color').text
    except:
        colour = ''

    
    try:
        item_bodytype = bs_item.find(itemprop='bodyType').text
    except:
        item_bodytype = ''


    try:
        configuration = bs_item.find(itemprop='vehicleConfiguration').text
    except:
        configuration = ''

    try:
        item_wheelConfig = bs_item.find(itemprop='driveWheelConfiguration').text
    except:
        item_wheelConfig = ''     

    try:
        item_price = bs_item.find('span',itemprop='price').text
        item_price = int(item_price.replace('$','').replace(',','').replace('.',' ').split()[0])
    except:
        item_price = 0

    try:
        item_transmission = bs_item.find(itemprop='vehicleTransmission').text
    except:
        item_transmission = ''


    try:
        item_mileage = bs_item.find(itemprop='mileageFromOdometer').text
        item_mileage = int(item_mileage.replace(',',''))
    except:
        item_mileage = 0
    


    item_carfax = bs_item.find('a', href=re.compile('^(https://reports.carproof.com)((?!:).)*$'))
    try:
        item_carfax_link = item_carfax.attrs['href']
    except:
        item_carfax = bs_item.find('a', href=re.compile('^(https://www.carproof.com)((?!:).)*$'))
        try: 
            item_carfax_link = item_carfax.attrs['href']
        except:
            item_carfax_link = ''

                 
    try:
        item_dealer_add = bs_item.find(itemprop='address').text
    except:
        item_dealer_add = ''
 

    

   
        

    
    every_element_list.append(brand)
    every_element_list.append(Model)
    every_element_list.append(date)
    every_element_list.append(item_price)
    every_element_list.append(colour)
    every_element_list.append(configuration)
    every_element_list.append(item_condition)
    every_element_list.append(item_bodytype)
    every_element_list.append(item_wheelConfig)
    every_element_list.append(item_transmission)
    every_element_list.append(item_fueltype)
    every_element_list.append(item_mileage)
    every_element_list.append(item_carfax_link)
    every_element_list.append(item_dealer_add)
   
    
    
    return every_element_list    
```

```python
big_data_list = []
itemlist = []
item_url = []

base_url='https://www.kijiji.ca'

number_of_pages=15000  
for page in range(1,number_of_pages):
    page_url = 'https://www.kijiji.ca/b-cars-vehicles/alberta/dodge__ford__gmc__hyundai__jeep__ram__toyota/'+'page+'+ str(page)+'/c174l9003a54'
    html = urlopen(page_url)
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^(/v-cars-trucks/)((?!:).)*$')):
        print('Page:',page)
        if 'href' in link.attrs:
            item_url = base_url + link.attrs['href']
            if '?' not in item_url:
                try:
                    print(item_url)
                    itemlist = get_item_info(item_url)
                    #print(itemlist)
                    big_data_list.append(itemlist)
                    itemlist = []
                    
                except:
                    pass
```

```python
BigDF = pd.DataFrame(big_data_list)
```
